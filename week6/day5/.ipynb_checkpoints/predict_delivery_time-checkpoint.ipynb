## 데이터 정제

### Rating column의 NaN과 숫자가 아닌 값을 평균값으로 채우기

```python
train_df.isnull().sum()
```

```python
train_num = train_df.copy()
```

```python
def non_numerals(series):
    non_numerals =[]
    for i in series.unique():
        try:
            i = float(i)
        except:
            non_numerals.append(i)
    return non_numerals

non_numerals(train_num['Rating'])
```

```python
train_num = train_num[(train_df['Rating']!='NEW')&(train_df['Rating']!='Opening Soon')&(train_df['Rating']!='Temporarily Closed')]
train_num
```

```python
# Rating column에서 NaN과 string으로 되어있는 값을 채워주기 위해 중앙값 확인

train_num['Rating'] = pd.to_numeric(train_num['Rating'])
train_num.describe()
```

```python
train_df.info()
```

```python
train_df['Rating'].fillna('3.6', inplace=True)
train_df.isnull().sum()
```

```python
train_df['Rating'] = train_df['Rating'].apply(lambda x: x.replace('NEW', '3.6'))
train_df['Rating'] = train_df['Rating'].apply(lambda x: x.replace('Opening Soon', '3.6'))
train_df['Rating'] = train_df['Rating'].apply(lambda x: x.replace('Temporarily Closed', '3.6'))
```

```python
train_df['Rating'] = pd.to_numeric(train_df['Rating'])
train_df.describe()
```

```python
train_df.isnull().sum()
```

### Votes와 Reviews의 누락된 값에 중간값으로 채우기

```python
train_df['Votes'].fillna(63, inplace=True)
train_df['Reviews'].fillna(26, inplace=True)
train_df.isnull().sum()
```

```python
# Votes column에서 숫자가 아닌 값이 있는 지 확인

def non_numerals(series):
    non_numerals =[]
    for i in series.unique():
        try:
            i = float(i)
        except:
            non_numerals.append(i)
    return non_numerals

non_numerals(train_df['Votes'])
```

```python
# Reviews column에서 숫자가 아닌 값이 있는 지 확인

def non_numerals(series):
    non_numerals =[]
    for i in series.unique():
        try:
            i = float(i)
        except:
            non_numerals.append(i)
    return non_numerals

non_numerals(train_df['Reviews'])
```

### Cuisines column one-hot encoding

```python
train_cat = train_df.copy()
train_cat
```

```python
onehot = train_cat['Cuisines'].str.get_dummies(sep=',')
onehot
```

```python
train_cat = pd.concat([train_cat, onehot], axis=1)
train_cat
```

```python
train_cat.drop(['Cuisines'], axis=1, inplace=True)
train_cat
```

```python
onehot_location = train_cat['Location'].str.get_dummies(sep=',')
onehot_location
```

```python
train_cat = pd.concat([train_cat, onehot_location], axis=1)
train_cat
```

```python
train_cat.drop(['Location'], axis=1, inplace=True)
train_cat
```

```python
# train_labels = train_cat['DeliveryTime'].copy()
# train_prepared = train_cat.drop(['Restaurant', 'DeliveryTime'], axis=1)
```

## 테스트 데이터셋 분리

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(train_cat, test_size=0.2, random_state=42)
```

```python
train_labels = train_set['DeliveryTime'].copy()
train_prepared = train_set.drop(['Restaurant', 'DeliveryTime'], axis=1)
```

## 모델 훈련 (Train a Model)

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(train_prepared, train_labels)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

delivery_predictions = lin_reg.predict(train_prepared)
lin_mse = mean_squared_error(train_labels, delivery_predictions)
lin_rmse = np.sqrt(lin_mse)
# lin_rmse
lin_mae = mean_absolute_error(train_labels, delivery_predictions)
lin_mae
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(train_prepared, train_labels)
```

```python
delivery_predictions = tree_reg.predict(train_prepared)
tree_mse = mean_squared_error(train_labels, delivery_predictions)
tree_rmse = np.sqrt(tree_mse)
# tree_rmse
tree_mae = mean_absolute_error(train_labels, delivery_predictions)
tree_mae
```

### 교차검증 (Cross-Validation)을 사용한 평가

결정트리 모델에 대한 평가

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, train_prepared, train_labels, scoring='neg_mean_squared_error', cv=10)
# scores_mae = cross_val_score(tree_reg, train_prepared, train_labels, scoring='mean_absolute_error', cv=10)

tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse_scores)
```

선형회귀 모델에 대한 평가

```python
lin_scores = cross_val_score(lin_reg, train_prepared, train_labels,
                             scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

RandomForestRegressor에 대한 평가

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=1000, random_state=42)
forest_reg.fit(train_prepared, train_labels)
```

```python
delivery_predictions = forest_reg.predict(train_prepared)
forest_mae = mean_absolute_error(train_labels, delivery_predictions)
# forest_rmse = np.sqrt(forest_mse)
forest_mae
```

```python
from sklearn.model_selection import cross_val_score

forest_mae_scores = cross_val_score(forest_reg, train_prepared, train_labels,
                                scoring="neg_mean_absolute_error", cv=10)
# forest_mae_scores = np.sqrt(-forest_scores)
display_scores(forest_mae_scores)
```

## 테스트 데이터셋으로 최종평가하기

```python
test_labels = test_set['DeliveryTime'].copy()
test_prepared = test_set.drop(['Restaurant', 'DeliveryTime'], axis=1)
```

렌덤포레스트 모델 평가

```python
delivery_predictions_test = forest_reg.predict(test_prepared)
forest_mae_test = mean_absolute_error(test_labels, delivery_predictions_test)
forest_mae_test
```

```python
under_prediction = test_labels-delivery_predictions_test

def ratio(x):
    under_prediction_count = 0
    for i in x:
        if i < 0:
            under_prediction_count += 1
    under_prediction_ratio = under_prediction_count / len(x)
    return under_prediction_ratio
    
ratio(under_prediction)
```

Linear Regression 모델 평가

```python
delivery_predictions_lin = lin_reg.predict(test_prepared)
lin_mae = mean_absolute_error(test_labels, delivery_predictions_lin)
lin_mae
```

```python
ratio(test_labels-delivery_predictions_lin)
```
