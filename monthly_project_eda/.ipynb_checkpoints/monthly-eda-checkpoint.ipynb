```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
income_df = pd.read_csv("./Korea Income and Welfare.csv")
income_df.head()
```

```python
income_df['occupation'].str.contains(' ')
income_df['occupation'] = income_df['occupation'].str.strip()
income_df
```

```python
def get_jobcode(x):
    if len(x) == 0:
        return 0
    elif len(x) > 3:
        return x[:2]
    else:
        return x[0]

income_df['job'] = income_df['occupation'].apply(lambda x: get_jobcode(x))
income_df
```

```python
clean_df = income_df.drop(income_df[income_df['job'].astype(int)==99].index)
clean_df
```

```python
plt.figure(figsize=(10, 6))
clean_df['job'] = pd.to_numeric(clean_df['job'])
clean_df.groupby('job', sort=True).mean()['income'].plot(kind='bar')
plt.title('Income-Job')
plt.ylabel('income')
plt.show()
```

```python
clean_df = clean_df.sort_values(by='year')
clean_df.groupby('year').mean()['income'].plot()
```

```python
keys = [year for year, df in clean_df.groupby('year')]

plt.figure(figsize=(10, 6))

sns.lineplot(x=keys, y=clean_df[clean_df['job']==1].groupby('year').mean()['income'])
sns.lineplot(x=keys, y=clean_df[clean_df['job']==2].groupby('year').mean()['income'])
sns.lineplot(x=keys, y=clean_df[clean_df['job']==3].groupby('year').mean()['income'])
# sns.lineplot(x=keys, y=clean_df[clean_df['job']==4].groupby('year').mean()['income'])
# sns.lineplot(x=keys, y=clean_df[clean_df['job']==5].groupby('year').mean()['income'])
# sns.lineplot(x=keys, y=clean_df[clean_df['job']==6].groupby('year').mean()['income'])
sns.lineplot(x=keys, y=clean_df[clean_df['job']==7].groupby('year').mean()['income'])
sns.lineplot(x=keys, y=clean_df[clean_df['job']==8].groupby('year').mean()['income'])
# sns.lineplot(x=keys, y=clean_df[clean_df['job']==9].groupby('year').mean()['income'])
sns.lineplot(x=keys, y=clean_df[clean_df['job']==10].groupby('year').mean()['income'])

plt.legend([1, 2, 3, 7, 8, 10])
plt.show()
```
